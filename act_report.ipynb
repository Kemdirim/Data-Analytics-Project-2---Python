## ANALYSIS AND VISUALIZATION OF WeRateDogs (@dog_rates) TWITTER DATA

## INSIGHTS AND VISUALISATION

After wrangling (gathering, assessing and cleaning) the data from the Twitter user @dog_rates, also known as **WeRateDogs** which is a Twitter account that rates people's dogs with a humorous comment about the dog, the following insights and visualisations were generated.

### INSIGHTS:

### 1. What source generated the most tweets? 

The above chart shows that most of the tweets contained in the dataset came from "Twitter for iPhone". In this study, all tweets originated from either of three sources namely,** TweetDeck, Twitter Web Client and Twitter for iPhone** with Twitter for iPhone accounting for about 98% of the tweets. This shows the dominance and efficiency of mobile usage in social media participation.

### 2. What dog stage was the most represented?

It is seen that the stages of the dogs in most of the tweets were not explicitly stated while some contained more than one dog stage, this will hinder the ability to make any conclusions or inferences. However from those availabale, the Pupper dog stage is found to be more represented.

### 3. What percentage/number of the first image prediction were identified dogs (p1_dog)?

The above chart shows that 74.07% (1477 of 1994) of the first image predictions were successfully identified as dogs and about 25.93% (517 of 1994) were not identified as dogs. Aalthough vetting the model used in this prediction isn't part of our task, we can however say that the model performed averagely and will probably need further tuning and training to improve its accuracy.

### VISUALIZATION:

### 1. Is there a correlation between the number of favorites (likes on the post) and the retweets generated? How are they related?

   The above scatter plot shows the relationship between the Favorite count (i.e the number of likes on a tweet post) and the Retweet count (i.e number of times a post was retweeted).

In a bid to understand the relationship or effect of the Favorite_count (number of likes), Retweet_count (number of retweets) and followers_count (number of followers of account that tweeted) on each other, their respective correlations were calculated:

**correlation:**\
 Favorite_count vs Retweet_count: 0.9129466274183703\
 Followers_count vs Retweet_count: -0.36427633697950684\
 Favorite_count vs Followers_count: -0.5480893568330488

As seen in the chart above, there is a positive correlation between the Favorite_count of a tweet and its Retweet_count with a correlation value of 0.9129466274183703. This shows that an increase in the number of likes on a tweet post, will most likely result in an increase in retweets.\
The correlation between Followers_count and Retweet_count as well as Favorite_count and Followers count are both negative, this definitely contradicts the popular opinion that the more followers gained would result in better engagement on tweets (likes on a post and retweets).
